```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_56.csv", skiprows=4)
```

```python
data
```

```python
selected_countries = data[data['Country Name'].isin(['India', 'China', 'United States', 'Brazil', 'Russia'])]
pop_2020 = selected_countries[['Country Name', '2020']].dropna()

plt.figure(figsize=(8, 6))
plt.bar(pop_2020['Country Name'], pop_2020['2020'], color='skyblue')
plt.title('Population in 2020 (Selected Countries)', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.xticks(rotation=45)
plt.show()

```

```python
pop_distribution_2020 = data[['Country Name', '2020']].dropna()

plt.figure(figsize=(10, 6))
plt.hist(pop_distribution_2020['2020'], bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Population Distribution in 2020 (All Countries)', fontsize=14)
plt.xlabel('Population', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
